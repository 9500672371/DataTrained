```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
link_request = requests.get('https://presidentofindia.nic.in/former-presidents.htm')
```

```python
link_request
```

```python
content = BeautifulSoup(link_request.content)
content
```

```python
Name = []
for i in content.find_all('div',class_="presidentListing"):
    Name.append(i.text.split('\n')[1])   
Name
```

```python
term = []
for i in content.find_all('div',class_="presidentListing"):
    term.append(i.text.split('\n')[2]) 
term
```

```python
len(Name),len(term)
```

```python
List=pd.DataFrame({'Name':Name,
                    'Term of Office':term,
                   })
List
```

```python
term_ = []
for i in List['Term of Office']:
    term_.append(i.split(':')[1]) 
term_
```

```python
List['Term of Office'] = term_
```

```python
List
```

```python
name_ = []
for i in List['Name']:
    name_.append(i.split(' (')[0]) 
name_
```

```python
List['Name'] = name_
List
```
